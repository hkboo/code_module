```python
def setup_mecab():
  !pip install jamo
  !pip install konlpy
  !git clone https://github.com/SOMJANG/Mecab-ko-for-Google-Colab.git
  %cd /content/Mecab-ko-for-Google-Colab
  !bash install_mecab-ko_on_colab_light_220111.sh
  %cd /content/mecab-ko-dic-2.1.1-20180720

  result = !ls user-dic
  if 'nnp.csv' in result:
    return True
  else:
    return False


def load_user_dic(path="./user-dic/nnp.csv"):
  with open(path, 'r', encoding='utf-8') as f:
    file_data = f.readlines()
  return file_data


def get_joungsung_TF(sample_text):
  from jamo import h2j, j2hcj

  sample_text_list = list(sample_text)
  last_word = sample_text_list[-1]
  last_word_jamo_list = list(j2hcj(h2j(last_word)))
  last_jamo = last_word_jamo_list[-1]
  jongsung_TF = "T"
  mo_l = "ㅏㅑㅓㅕㅗㅛㅜㅠㅡㅣㅘㅚㅙㅝㅞㅢㅐㅔㅟㅖㅒ"
  if last_jamo in [m for m in mo_l]:
    jongsung_TF = "F"
  return jongsung_TF


def create_add_dict(word_list):
  def create_line(word):
    jongsung_TF = get_joungsung_TF(word)
    line = '{},,,,NNP,*,{},{},*,*,*,*,*\n'.format(word, jongsung_TF, word)
    return line

  user_dic_l = []
  for word in word_list:
    user_dic_l.append(create_line(word))

  return user_dic_l


def add_and_save_user_words(user_dic_l, path="./user-dic/nnp.csv"):
  file_data = load_user_dic(path=path)
  cols = file_data[0]
  file_data += user_dic_l
  file_data = list(set(file_data))
  with open(path, 'w', encoding='utf-8') as f:
    for line in file_data:
      f.write(line)

  return


def run_add_userdict_shell():
  !bash ./tools/add-userdic.sh
  !make install
  return


# 메캅 셋업
setup_mecab()

# 정상 동작 확인
from konlpy.tag import Mecab

mecab = Mecab()
print(1, mecab.pos("네이버 스토어팜 부현경"))

# 추가할 사용자 단어 지정
word_list = ['네이버', '스토어', '세트', '부현경']

# 추가한 사용자 단어 추가
user_dic_l = create_add_dict(word_list)
add_and_save_user_words(user_dic_l)
print(load_user_dic())

# 추가된 사용자 사전 인스톨
run_add_userdict_shell()
```
